```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

queimadas_brasil_2024['geometry'] = gpd.points_from_xy(queimadas_brasil_2024['lon'], queimadas_brasil_2024['lat'])

queimadas_brasil_2024 = gpd.GeoDataFrame(queimadas_brasil_2024, geometry='geometry', crs="EPSG:4326")

# Reprojetar os dados para o CRS adequado para Contextily (EPSG:3857)
queimadas_brasil_2024 = queimadas_brasil_2024.to_crs(epsg=3857)

# Plotar o mapa
ax = queimadas_brasil_2024.plot(figsize=(15, 10), markersize=2, color='red', alpha=0.6)

ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

# Configurar título e ocultar os eixos
plt.title('Focos de Queimadas no Brasil em Setembro de 2024')
ax.set_axis_off()

# Mostrar o mapa
plt.show()
```

```python
import matplotlib.pyplot as plt
import contextily as ctx

# Reprojetar os dados para o CRS adequado para Contextily (EPSG:3857)
queimadas_brasil_2024 = queimadas_brasil_2024.to_crs(epsg=3857)

# Plotar o mapa
ax = queimadas_brasil_2024.plot(figsize=(15, 10), markersize=2, color='red', alpha=0.6)

# Use an alternative tile provider, like OpenStreetMap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)
# This line is now uncommented to use OpenStreetMap instead of the deprecated Stamen TonerLite

# Configurar título e ocultar os eixos
plt.title('Focos de Queimadas no Brasil em Setembro de 2024')
ax.set_axis_off()

# Mostrar o mapa
plt.show()
```

```python
plt.figure(figsize=(12,5))
ax = sns.countplot(x='bioma', data=queimadas_brasil_2024, palette='inferno')
plt.title('Quantidade de queimadas por bioma')
ax.set_xlabel('Bioma')
ax.set_ylabel('Qtd. de queimadas')
plt.show()
```

```python
# prompt: Gere no mesmo gráfico a região mais afetada apresentando as cidades/estado que mais registraram queimadas nesse dataset, pode mostrar o top 10 em cada quesito

# Top 10 estados com mais queimadas
top_10_estados = queimadas_brasil_2024.groupby('estado')['estado'].count().sort_values(ascending=False).head(10)

# Top 10 cidades com mais queimadas
top_10_cidades = queimadas_brasil_2024.groupby('municipio')['municipio'].count().sort_values(ascending=False).head(10)

# Criar o gráfico
plt.figure(figsize=(15, 8))

# Gráfico de barras para os estados
plt.subplot(1, 2, 1)
plt.bar(top_10_estados.index, top_10_estados.values)
plt.title('Top 10 Estados com mais Queimadas')
plt.xlabel('Estado')
plt.ylabel('Número de Queimadas')
plt.xticks(rotation=45, ha='right')

# Gráfico de barras para as cidades
plt.subplot(1, 2, 2)
plt.bar(top_10_cidades.index, top_10_cidades.values)
plt.title('Top 10 Cidades com mais Queimadas')
plt.xlabel('Cidade')
plt.ylabel('Número de Queimadas')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
