```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Assuming 'lat' and 'lon' columns exist in your DataFrame
# Create a geometry column from the latitude and longitude
queimadas_brasil_2024['geometry'] = gpd.points_from_xy(queimadas_brasil_2024['lon'], queimadas_brasil_2024['lat'])

# Convert the Pandas DataFrame to a GeoDataFrame
# Instead of EPSG:404000, use a valid CRS, for example, EPSG:4326 (WGS 84)
queimadas_brasil_2024 = gpd.GeoDataFrame(queimadas_brasil_2024, geometry='geometry', crs="EPSG:4326")

# Reprojetar os dados para o CRS adequado para Contextily (EPSG:3857)
queimadas_brasil_2024 = queimadas_brasil_2024.to_crs(epsg=3857)

# Plotar o mapa
ax = queimadas_brasil_2024.plot(figsize=(15, 10), markersize=2, color='red', alpha=0.6)

# Use an alternative tile provider, like OpenStreetMap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

# Configurar título e ocultar os eixos
plt.title('Focos de Queimadas no Brasil em Setembro de 2024')
ax.set_axis_off()

# Mostrar o mapa
plt.show()
```

```python
plt.figure(figsize=(12,5))
ax = sns.countplot(x='bioma', data=queimadas_brasil_2024, palette='inferno')
plt.title('Quantidade de queimadas por bioma')
ax.set_xlabel('Bioma')
ax.set_ylabel('Qtd. de queimadas')
plt.show()
```
